#### <font color='Purple'> Histogram </font>
**Histogram** represents the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.

* Function `distplot()` provides the most convenient way to take a quick look at univariate distribution. 

`seaborn.distplot()`
    * Always for numerical data

    data - Series, 1d array or a list	
    bins - Specification of hist bins	
    hist - bool
    kde - bool

```python
df = sns.load_dataset('iris')
df
```

#### <font color='Purple'> Relplot </font>
**Relplot** or relation plot between variables.
* Numerical to Numerical plot

    * scatterplot() (with kind=”scatter”) (default)
    * lineplot() (with kind=”line”)


```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind = 'scatter')
```

```python
#. Positive co relation. Total bill is increasing , tip will also increase. 
#. Density is more between 10 - 20. 
#. Data points are more sparse beyond bill 25. 
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="line")
```

```python
# Adding hue 
    # Relationship every day of week 
    # Always is categorical variable. 
    
sns.relplot(data=tips, x="total_bill", y="tip", hue = 'day')
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col = 'day', col_wrap = 2)
```

```python
# Adding hue and col 
    # Analyzing different day separately
    
sns.relplot(data=tips, x="total_bill", y="tip", hue="sex", col="day", col_wrap=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", col="day", col_wrap=2)
```

```python
sns.relplot(data=tips, x="size", y="tip",kind="line")
```

##### Seaborn - Distribution of Observations

* Distribution of values within each category. 

`boxplot()` 
* Categorical to Numerical graph

**Boxplot** is a convenient way to visualize the distribution of data through their quartiles.
* Box plots usually have vertical lines extending from the boxes which are termed as whiskers. 
* These whiskers indicate variability outside the upper and lower quartiles, hence Box Plots are also termed as box-and-whisker plot and box-and-whisker diagram. 
* Any Outliers in the data are plotted as individual points.

```python
df = sns.load_dataset('iris')
sns.boxplot(x = "species", y = "petal_length", data = df)
plt.show()

## dots are outliers. 
```

```python
df = sns.load_dataset('tips')
sns.boxplot(x = "day", y = "total_bill", data=df)
plt.show()
```

```python
# Adding hue

df = sns.load_dataset('tips')
sns.boxplot(x = "day", y = "total_bill", hue = 'sex', data=df)
plt.show()
```

`violinplot()` 
* Numerical to Categorical

**Violinplot** is a combination of the box plot with the kernel density estimates.


```python
df = sns.load_dataset('tips')
sns.violinplot(x = "day", y = "total_bill", data=df)
plt.show()
```

```python
df = sns.load_dataset('iris')
sns.violinplot(x = "species", y = "petal_length", data=df)
plt.show()
```

* The wider portion of violin indicates the higher density and narrow region represents relatively lower density. 
* The Inter-Quartile range in boxplot and higher density portion in kde fall in the same region of each category of violin plot.

```python
# Above relationship with respect to sex - adding hue

df = sns.load_dataset('tips')
sns.violinplot(x = "day", y = "total_bill",hue = 'sex', data = df)
plt.show()
```

```python
# If hue has only 2 class, we can split the same plot into two insteead of 2 diff violins

df = sns.load_dataset('tips')
sns.violinplot(x = "day", y = "total_bill",hue = 'sex',split= True, data = df)
plt.show()
```

##### Seaborn - Statistical Estimation

* Estimation of Central limit theorem. 

`barplot()` 

**Barplot** is a relation between a categorical variable and a continuous variable
* The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.barplot(x = "sex", y = "survived", hue = "class", data = df)
plt.show()
```

```python
# Horizontal barplot

sns.barplot(x = 'age', y = 'embark_town', data = df,
            palette = 'PuRd', orient = 'h',
            )
plt.show()
```

`Countplot()` 
* Only takes one category

**Countplot** is special case in barplot is to show the no of observations in each category rather than computing a statistic for a second variable

```python
sns.countplot(x = "class", data = df, palette = "Blues");
plt.show()
```

```python
# Adding count number on top

ax = sns.countplot(x = "class", data = df, palette = "Blues");
ax.bar_label(ax.containers[0])
```

```python
sns.countplot(x = 'class', hue = 'who', data = df, palette = 'magma',)
plt.title('Survivors')
plt.show()
```

```python
ax = sns.countplot(x = 'class', hue = 'who', data = df, palette = 'magma')
for container in ax.containers:
    ax.bar_label(container)
```

`Pointplot()` 

**Pointplot** is  used to show point estimates and confidence intervals using scatter plot glyphs.
* A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.
* Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables. 

```python
df = sns.load_dataset('titanic')
sns.pointplot(x = "sex", y = "survived", hue = "class", data = df)
plt.show()
```

```python
data = sns.load_dataset("tips") 
sns.pointplot(x="day", y="tip", data=data)
plt.show()
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=data, palette="Accent")
```

##### Seaborn - Linear Relationships
* To check linear relationships across multiple variables. 
* `regplot()` and `lmplot()`

`regplot()` 

**Regplot** accepts the x and y variables in a variety of formats including simple numpy arrays, pandas Series objects, or as references to variables in a pandas DataFrame

`lmplot()`

**Lmplot** :: Regplot plus extension of facet grid



```python
tips = sns.load_dataset("tips")
ax = sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
## We can also add some jitters around 

sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
sns.regplot(x="total_bill", y="tip", data=tips, x_jitter=1, ci = None)  #Removes the shaded portion
```

`lmplot()`

* Basic & advanced cases

```python
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="day", hue="day",
               data=tips, col_wrap=2, height=3)
```

#### HeatMap

* A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours.

```python
flights=sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
flights
```

```python
sns.heatmap(flights,cmap="RdBu")
```

```python
sns.heatmap(flights,cmap="Greens",linewidths=.5) # more friendly
```

```python
## Also useful to plot correlations 

car_crashes = sns.load_dataset("car_crashes")
corr=car_crashes.corr()

corr
```

```python
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGnBu")
```

`Pairplot()`

* It creates a grid of Axes such that each numeric variable in data will be shared across the y-axes across a single row and the x-axes across a single column. 
* The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.
* Pair Plots are a really simple way to visualize relationships between each variable

```python
df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
